## Merge la School_Details & Levels_Formation

```python
# Incarcare csv-uri
LF = pd.read_csv("Levels_Formation_Clean_Team4.csv")
school_details = pd.read_csv("data.csv")
```

```python
# Cu merge
all_details = pd.merge(LF, school_details, on = "idSchool", how = "inner")
```

```python
all_details
```

```python
all_details["Limba_de_predare"].unique()
```

```python
# Am calculat numarul total de scoli din fiecare judet

total_schools_count = all_details.groupby('Judet')['idSchool'].count().reset_index()
total_schools_count.columns = ['Judet', "Numar_total_scoli"]
```

```python
# Am impartit coloana "Limba_de_predare" in coloane separate pentru fiecare limba de predare in parte
all_details = pd.get_dummies(all_details, columns = ['Limba_de_predare'], prefix = ' ')
```

```python
#Reordonez totul in functie de 'Judet' si folosesc functia sum() pentru a calcula numarul de scoli care au o anumita limba de predare 

all_details_grouped = all_details.groupby("Judet").sum().reset_index()
```

```python
#Dupa ce setul de date este reordonat in functie de 'Judet', si, avand in vedere ca in coloana "Nivel" valorile din randuri 
#se repetau, am folosit o functie pentru a selecta doar prima "valoare" identificata in acea repetare, in cazul de fata "Liceu"

nivel = all_details.groupby('Judet')['Nivel'].agg(lambda x: x.iloc[0]).reset_index()
```

```python
#Aduc totul impreuna intr-un singur df
result = pd.merge(total_schools_count, all_details_grouped, on = 'Judet')
```

```python
final = pd.merge(result, nivel, on = 'Judet')
```

```python
final
```

```python
#Elimin coloanele nefolositoare
final.drop(columns = ['idSchool', 'Nivel_x'], inplace=True)
```

```python
final.columns
```

```python
#Redenumesc coloanele
final.rename(columns = {' _Limba germană':"liceu_german", ' _Limba maghiară':'liceu_maghiar', ' _Limba polonă':'liceu_polonez', 
                        ' _Limba română': 'liceu_romanesc',' _Limba ucraineană':'liceu_ucrainean', 'Nivel_y':'Nivel' }, inplace=True)
```
