### GET LINK FUNCTION

```python
def get_link (x, code):
    for i in x:
        if i["code"] == code:
            website = i["value"]
            return website
    return np.nan        
```

```python
get_link(df_withlink.schoolSocialLinks.iloc[80], "WEB" )
```

```python
details_df["Web_Link"] = details_df.schoolSocialLinks.apply(lambda x: get_link(x, 'WEB'))
```

```python
details_df
```

```python
details_df["TWI_Link"] = details_df.schoolSocialLinks.apply(lambda x: get_link(x, 'TWI'))
```

```python
details_df["FB_Link"] = details_df.schoolSocialLinks.apply(lambda x: get_link(x, 'FB'))
```
